Task about lakes

Generate 20 random variable from 1 to 100

Draw the plot of the sequence

Lets assume each point represent the height and so all plotting is 2d mountains.

Then consider the unlimited rain from above - cavities become lakes full of water.

Determine the deepest lake

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерація масиву з 20 випадкових значень
values = np.random.randint(0, 101, 20)
print("Generated values:", values)

# Функція для обчислення глибини найглибшого озера
def calculate_deepest_lake(values):
    max_depth = 0
    n = len(values)
    
    for i in range(1, n - 1):
        if values[i - 1] > values[i] < values[i + 1]:
            left, right = i - 1, i + 1
            
            while left > 0 and values[left - 1] >= values[left]:
                left -= 1
            while right < n - 1 and values[right + 1] >= values[right]:
                right += 1
            
            lake_depth = min(values[left], values[right]) - values[i]
            max_depth = max(max_depth, lake_depth)
    
    return max_depth

# Обчислення глибини найглибшого озера
deepest_lake_depth = calculate_deepest_lake(values)
print("Deepest lake depth:", deepest_lake_depth)

# Побудова графіка
plt.plot(values, marker='o', linestyle='-', color='b')

# Виділення найглибшого озера
for i in range(1, len(values) - 1):
    if values[i - 1] > values[i] < values[i + 1]:
        left, right = i - 1, i + 1
        
        while left > 0 and values[left - 1] >= values[left]:
            left -= 1
        while right < len(values) - 1 and values[right + 1] >= values[right]:
            right += 1
        
        lake_depth = min(values[left], values[right]) - values[i]
        if lake_depth == deepest_lake_depth:
            plt.plot(np.arange(left, right + 1), values[left:right + 1], marker='o', linestyle='-', color='red')

plt.xlabel('Position')
plt.ylabel('Height')
plt.show()
```
